```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  # Output: 120
```

```python
def count_up(n):
    for i in range(1, n + 1):
        yield i
list(count_up(5))
```

```python
def evens():
    num = 0
    while True:
        yield num
        num += 2
import itertools
list(itertools.islice(evens(),5))
```

```python
def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
list(itertools.islice(fib(),10))
```

```python
import time
from functools import wraps

def timed(fn):
    @wraps(fn)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = fn(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"Function {fn.__name__} took {elapsed:.6f} seconds")
        return result
    return wrapper

@timed
def slow_add(a, b):
    time.sleep(0.1)
    return a + b

slow_add(2, 3)
```

```python
# Context manager via function (contextlib)
from contextlib import contextmanager

@contextmanager
def open_upper(path):
    f = open(path, "r")
    try:
        yield (line.upper() for line in f)
    finally:
        f.close()

# with open_upper("file.txt") as lines:
#     for line in lines: ...
```

```python
# Pattern matching in functions (Python 3.10+)
def describe(value):
    match value:
        case 0:
            return "zero"
        case 1 | 2:
            return "one or two"
        case [x, y]:
            return f"list of two: {x}, {y}"
        case _:
            return "something else"

describe([3, 4])  # "list of two: 3, 4"
```

```python
# Type hints + docstring
from typing import Iterable

def avg(nums: Iterable[float]) -> float:
    """
    Return the arithmetic mean of a non-empty iterable of numbers.
    """
    nums = list(nums)
    return sum(nums) / len(nums)

avg([1, 2, 3, 4])
```

```python
# Memoization of expensive function
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n: int) -> int:
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

fib(10)  # 55

```

```python
# Higher-order decorator factory with parameters
def repeat(times=2):
    def decorator(fn):
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(times):
                result = fn(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(times=3)
def shout(msg):
    print(msg.upper())

shout("hi")
```
