```python
student = {
    "name": "Balaji",
    "age": 40,
    "skill": "Python"
}

keys = list(student.keys())
i = 0

while i < len(keys):
    key = keys[i]
    print(key, ":", student[key])
    i += 1
```

```python
marks = {"math": 90, "english": 85, "science": 92}

subjects = list(marks.keys())
i = 0
total = 0

while i < len(subjects):
    total += marks[subjects[i]]
    i += 1

print("Total marks =", total)
print("Average marks =", total / len(subjects))
```

```python
data = [[1, 2, 3], [4, 5, 6], [7, 8]]
i = 0
sum_even = 0

while i < len(data):
    j = 0
    while j < len(data[i]):
        if data[i][j] % 2 == 0:
            sum_even += data[i][j]
        j += 1
    i += 1

print("Sum of even numbers =", sum_even)
```

```python
i = 1
while i <= 10:
    if i == 5:
        break   # loop stops completely
    print(i)
    i += 1
```

```python
i = 1
while i <= 10:
    if i == 5:
        i += 1      # important to avoid infinite loop
        continue    # skip printing 5
    print(i)
    i += 1
```

```python
numbers = [3, 7, 9, 12, 15]
i = 0

while i < len(numbers):
    if numbers[i] % 2 == 0:
        print("First even number =", numbers[i])
        break
    i += 1
```

### While loop with continue - print only odd numbers

```python
i = 0

while i <= 10:
    if i % 2 == 0:
        i += 1
        continue   # skip even numbers
    print(i)
    i += 1
```

```python
my_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
vowels = ['a', 'e', 'i', 'o', 'u']
i = 0

while i < len(my_list):
    if my_list[i] in vowels:
        i += 1
        continue
    print(my_list[i], "is a consonant")
    i += 2
```
