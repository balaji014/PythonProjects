```python
import time
from functools import wraps

def timed(fn):
    @wraps(fn)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = fn(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"Function {fn.__name__} took {elapsed:.6f} seconds")
        return result
    return wrapper

@timed
def slow_add(a, b):
    time.sleep(0.1)
    return a + b

slow_add(2, 3)
```

```python
# Context manager via function (contextlib)
from contextlib import contextmanager

@contextmanager
def open_upper(path):
    f = open(path, "r")
    try:
        yield (line.upper() for line in f)
    finally:
        f.close()

# with open_upper("file.txt") as lines:
#     for line in lines: ...
```

```python
# Pattern matching in functions (Python 3.10+)
def describe(value):
    match value:
        case 0:
            return "zero"
        case 1 | 2:
            return "one or two"
        case [x, y]:
            return f"list of two: {x}, {y}"
        case _:
            return "something else"

describe([3, 4])  # "list of two: 3, 4"
```

```python
# Type hints + docstring
from typing import Iterable

def avg(nums: Iterable[float]) -> float:
    """
    Return the arithmetic mean of a non-empty iterable of numbers.
    """
    nums = list(nums)
    return sum(nums) / len(nums)

avg([1, 2, 3, 4])
```

```python
# Memoization of expensive function
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n: int) -> int:
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

fib(10)  # 55

```

```python
# Higher-order decorator factory with parameters
def repeat(times=2):
    def decorator(fn):
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(times):
                result = fn(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(times=3)
def shout(msg):
    print(msg.upper())

shout("hi")
```

```python
# Function with docstring + type hints
from typing import Iterable

def product(nums: Iterable[int]) -> int:
    """
    Multiply all numbers together; returns 1 for an empty iterable.
    """
    result = 1
    for n in nums:
        result *= n
    return result
```

```python
# Using nonlocal (modifying outer-scope variable in nested function)
def make_counter():
    count = 0
    def inc():
        nonlocal count
        count += 1
        return count
    return inc

c = make_counter()
c(), c(), c()  # 1, 2, 3
```

```python
# Recursive with memoization
from functools import lru_cache

@lru_cache(maxsize=None)
def climb_stairs(n: int) -> int:
    if n <= 2:
        return n
    return climb_stairs(n - 1) + climb_stairs(n - 2)

climb_stairs(5)  # 8
```

```python
# Curried function (returns another configured function)
def raise_to(power):
    def inner(x):
        return x ** power
    return inner

square = raise_to(2)
cube = raise_to(3)
square(5), cube(2)  # (25, 8)
```

```python
# Decorator that enforces types at runtime (lightweight check)
def enforce_ints(fn):
    def wrapper(*args):
        if not all(isinstance(a, int) for a in args):
            raise TypeError("all arguments must be int")
        return fn(*args)
    return wrapper

@enforce_ints
def add_three(a, b, c):
    return a + b + c

add_three(1, 2, 3)
# add_three(1, "x", 3)  # raises TypeError
```

```python
# 1) Nested dict transformation
def normalize_users(users: dict) -> dict:
    # Lower-case emails and ensure 'active' flag
    return {
        uid: {**info, "email": info["email"].lower(), "active": info.get("active", True)}
        for uid, info in users.items()
    }

users = {
    "u1": {"name": "Ava", "email": "Ava@Example.com"},
    "u2": {"name": "Ben", "email": "ben@EXAMPLE.com", "active": False},
}
normalize_users(users)
```

```python
# 2) List of tuples â†’ grouped dict
from collections import defaultdict

def group_scores(pairs: list[tuple[str, int]]) -> dict[str, list[int]]:
    grouped: dict[str, list[int]] = defaultdict(list)
    for name, score in pairs:
        grouped[name].append(score)
    return dict(grouped)

group_scores([("amy", 10), ("bob", 8), ("amy", 7)])
```

```python
# 3) Set operations in a helper
def common_unique(a: set[int], b: set[int]) -> tuple[set[int], set[int], set[int]]:
    return a & b, a - b, b - a  # intersection, only-in-a, only-in-b

common_unique({1, 2, 3}, {3, 4, 5})
```

```python
# 4) Data class with method
from dataclasses import dataclass

@dataclass
class Order:
    id: int
    items: list[tuple[str, float]]  # (name, price)

    def total(self) -> float:
        return sum(price for _, price in self.items)

order = Order(1, [("apple", 1.2), ("bread", 2.5)])
order.total()
```

```python
# 5) Typed matrix operations
from typing import Sequence

Matrix = Sequence[Sequence[float]]

def transpose(m: Matrix) -> Matrix:
    # Assumes all rows are equal length
    return [list(row) for row in zip(*m)]

transpose([[1, 2, 3], [4, 5, 6]])
```
