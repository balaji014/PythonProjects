```python
def greet():
    print("Hello, World!")
greet()
```

```python
def add(a, b):
    print(a + b)
add(3, 5)
```

```python
def multiply(x, y):
    return x * y
result = multiply(4, 6)
print(result)
```

```python
def welcome(name="Guest"):
    print(f"Welcome, {name}!")
welcome()
welcome("Balaji")
welcome("Jhansi")
```

```python
def person(name, age):
    print(name,age)
person(age=40, name="Balaji")
person(name="Jhansi", age=10)
```

```python
def total_sum(*numbers):
    sum = 0
    for num in numbers:
        sum += num
    return sum
result = total_sum(1, 2, 3, 4, 5, 10)
print(result)
```

```python
def profile(**details):
    print(details)
profile(name="Balaji", age=40, role="Architect")
```

```python
def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers)/len(numbers)
mn, mx, avg = get_stats([10, 20, 30, 40, 50])
print(mn, mx, avg)
```

```python
def outer():
    print("Outer function")

    def inner():
        print("Inner function")
    inner()
outer()
```

```python
def apply(func, value):
    return func(value)
def square(x):
    return x * x
result = apply(square, 7)
print(result)
```
